# Exercise 2: Analytical Solution of Phytoplankton and Nutrient Model
This notebook plots the analytical solution of the model presented in equation (4) to show the time evolution of phytoplankton (P) and nutrient (N) concentrations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k_prime = 0.1  # rate constant (per day per concentration)
P0 = 0.1       # initial phytoplankton concentration (mmol N m^-3)
N0 = 5.0       # initial nutrient concentration (mmol N m^-3)
t = np.linspace(0, 20, 200)  # time axis in days

# Analytical solutions
# P(t) = (P0 * (N0 + P0)) / ((P0 + N0 * exp(-k_prime * (N0 + P0) * t))
# N(t) = (N0 * (N0 + P0)) / (P0 * exp(k_prime * (N0 + P0) * t) + N0)

denominator = P0 + N0 * np.exp(-k_prime * (N0 + P0) * t)
P_t = P0 * (N0 + P0) / denominator
N_t = N0 * (N0 + P0) / (P0 * np.exp(k_prime * (N0 + P0) * t) + N0)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, P_t, label='Phytoplankton (P)')
plt.plot(t, N_t, label='Nutrient (N)')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (mmol N m$^{-3}$)')
plt.title('Analytical Solution of Phytoplankton and Nutrient Model')
plt.legend()
plt.show()
```
